```python
for col in data.columns:
    if col.startswith('left') or col.startswith('right'):
        new_col_name = f"{col}_binary"
        data[new_col_name] = np.where(data[col]>0, 1, 0)
```

```python

df = data.copy(deep=True)

# Suponiendo que df es tu DataFrame original

# Paso 1: Calcular las influencias de cada variable binaria sobre cada expresión
influences = {}
for exp in df.ground_truth.unique():
    df2 = df[df.ground_truth == exp]
    binary_influence = {}
    binary_columns = [col for col in df2.columns if col.endswith('_binary')]
    for col in binary_columns:
        binary_influence[col] = df2[col].mean()
    influences[exp] = binary_influence

# Paso 2: Convertir el diccionario de influencias en un DataFrame
influences_df = pd.DataFrame(influences).T

# Paso 3: Graficar las influencias
plt.figure(figsize=(14, 8))
influences_df.plot(kind='bar', figsize=(15, 8), width=0.8)

# Ajustes de la gráfica
plt.title("Influencia de Variables Binarias sobre Cada Expresión en Ground Truth")
plt.xlabel("Expresión")
plt.ylabel("Influencia (Media de Variables Binarias)")
plt.legend(title="Variables Binarias", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que df es tu DataFrame original
# Paso 1: Calcular las influencias de cada variable binaria sobre cada expresión
influences_left = {}
influences_right = {}
for exp in df.ground_truth.unique():
    df2 = df[df.ground_truth == exp]
    binary_influence_left = {}
    binary_influence_right = {}
    binary_columns = [col for col in df2.columns if col.endswith('_binary')]
    for col in binary_columns:
        if 'left' in col:
            binary_influence_left[col] = df2[col].mean()
        elif 'right' in col:
            binary_influence_right[col] = df2[col].mean()
    influences_left[exp] = binary_influence_left
    influences_right[exp] = binary_influence_right

# Paso 2: Convertir los diccionarios de influencias en DataFrames
influences_left_df = pd.DataFrame(influences_left).T
influences_right_df = pd.DataFrame(influences_right).T

# Paso 3: Graficar las influencias de los canales "left"
plt.figure(figsize=(14, 8))
influences_left_df.plot(kind='bar', figsize=(15, 8), width=0.8)
plt.title("Influencia de Variables Binarias (Left) sobre Cada Expresión en Ground Truth")
plt.xlabel("Expresión")
plt.ylabel("Influencia (Media de Variables Binarias)")
plt.legend(title="Variables Binarias (Left)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Paso 4: Graficar las influencias de los canales "right"
plt.figure(figsize=(14, 8))
influences_right_df.plot(kind='bar', figsize=(15, 8), width=0.8)
plt.title("Influencia de Variables Binarias (Right) sobre Cada Expresión en Ground Truth")
plt.xlabel("Expresión")
plt.ylabel("Influencia (Media de Variables Binarias)")
plt.legend(title="Variables Binarias (Right)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Comparison of Left and Right EEG Channel Influences on Emotional Expressions

### 1. General Activation Patterns Across Emotions

- **Joy and Surprise**: 
  - **Left Side**: Channels on the left side showed significant activation for "joy" and "surprise," with prominent influence from channels like `left_F7_imp_binary` and `left_Fp1_imp_binary`.
  - **Right Side**: The right side also showed considerable activation for "surprise," but "joy" had relatively lower activation in comparison to the left. Channels like `right_F8_imp_binary` and `right_Fp2_imp_binary` were particularly active for "surprise."

- **Sad and Neutral**:
  - **Left Side**: "Sad" had a lower influence overall across left channels, indicating minimal activation for this emotion. The "neutral" expression also showed relatively lower activation in left-side channels.
  - **Right Side**: "Neutral" had a higher influence in right channels than in the left, with channels like `right_F8_imp_binary` and `right_Fp2_imp_binary` showing notable activity. "Sad" had a more distributed pattern on the right, showing moderate activation across multiple channels.

- **Angry**:
  - **Left Side**: The influence for "angry" on the left side showed moderate activation, with no single channel showing dominant influence.
  - **Right Side**: The activation pattern for "angry" on the right side was similar, with moderate influence across several channels, indicating a balanced pattern without prominent peaks in either hemisphere.

### 2. Channel-Specific Activation Differences

- **Frontal Channels (e.g., F7/Fp1 on the left and F8/Fp2 on the right)**:
  - **Left Frontal Channels**: The left frontal channels, particularly `left_F7_imp_binary` and `left_Fp1_imp_binary`, showed strong influence for emotions like "joy" and "surprise," indicating the left hemisphere's role in processing positive emotions.
  - **Right Frontal Channels**: The right frontal channels, such as `right_F8_imp_binary` and `right_Fp2_imp_binary`, exhibited more influence for "surprise" and "neutral." This aligns with the right hemisphere's role in processing complex and potentially negative emotions.

- **Distributed Channels (C and P Regions)**:
  - On both sides, channels in the C and P regions (e.g., `C3`, `C4`, `P3`, `P4`) showed moderate influence for emotions across the board. These channels were not as strongly associated with a specific emotion as the frontal channels, reflecting a more balanced or auxiliary role in emotional processing.

### 3. Hemispheric Differences in Emotional Processing

- **Positive Emotions (Joy)**: The left hemisphere appears to have a stronger role in processing positive emotions such as "joy," as evidenced by the higher influence in left channels compared to right.
  
- **Negative and Complex Emotions (Surprise, Neutral)**: The right hemisphere shows a stronger response to complex or mixed-affect emotions like "surprise" and "neutral." This is consistent with research suggesting the right hemisphere is more involved in processing complex or less clearly defined emotions, which might require more nuanced cognitive processing.

### 4. Conclusion

- **Left Hemisphere**: Shows greater influence for positive emotions (particularly "joy") and emotions involving approach-related processing. Channels like `left_F7_imp_binary` and `left_Fp1_imp_binary` are especially responsive.
- **Right Hemisphere**: Shows greater influence for complex and neutral emotions, with channels like `right_F8_imp_binary` and `right_Fp2_imp_binary` being particularly responsive. The right hemisphere might play a more active role in processing emotions with ambiguous valence or those that may require more cognitive evaluation.

This comparison highlights the complementary roles of the left and right hemispheres in emotional processing, with the left hemisphere showing more activation for positive emotions and the right hemisphere showing more activation for complex or mixed emotions. This aligns with the lateralization of emotional processing in the brain, where the left hemisphere is often associated with approach-related emotions and the right hemisphere with withdrawal or complex affective states.
